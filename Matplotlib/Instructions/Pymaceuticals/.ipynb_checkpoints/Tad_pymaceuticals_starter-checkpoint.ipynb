```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
trial_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
merged_df = trial_data.merge(mouse_data, on="Mouse ID")
# Or merged_df = trial_data.merge(mouse_data, on = 'Mouse ID')

# Display the data table for preview
merged_df.head(20)
```

## Tumor Response to Treatment

```python
# Select columns of interest
tumor_response = merged_df[['Timepoint', 'Tumor Volume (mm3)', 'Drug']]
tumor_response.head()
```

```python
# Group by Drug and Timepoint. Get average Tumor response to treatment.
tumor_response = pd.DataFrame(tumor_response.groupby(by=['Drug', 'Timepoint'])['Tumor Volume (mm3)'].mean())
tumor_response.head()
```

```python
# Calculate standard error for Tumor Volume
tumor_response['Volume SEM'] = tumor_response['Tumor Volume (mm3)'].sem()
tumor_response.head()
```

```python
tumor_response_pivot = pd.pivot_table(data=merged_df, index='Timepoint', columns='Drug', values='Tumor Volume (mm3)',
                               aggfunc = np.mean)
tumor_response_pivot.head(10)
```

**Shows tumor reactions among different drugs**


```python
tumor_response_sem_pivot = tumor_response.pivot_table(index='Timepoint',columns='Drug',values='Volume SEM')  
tumor_response_sem_pivot
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(tumor_response_pivot.index, tumor_response_pivot['Capomulin'], yerr=tumor_response_sem_pivot['Capomulin'],
            color='r', marker='o', markersize=5, linestyle='--', linewidth=0.5)
plt.errorbar(tumor_response_pivot.index, tumor_response_pivot['Infubinol'], yerr=tumor_response_sem_pivot['Infubinol'],
            color='b', marker='^', markersize=5, linestyle='--', linewidth=0.5)
plt.errorbar(tumor_response_pivot.index, tumor_response_pivot['Ketapril'], yerr=tumor_response_sem_pivot['Ketapril'],
            color='g', marker='s', markersize=5, linestyle='--', linewidth=0.5)
plt.errorbar(tumor_response_pivot.index, tumor_response_pivot['Placebo'], yerr=tumor_response_sem_pivot['Placebo'],
            color='black', marker='D', markersize=5, linestyle='--', linewidth=0.5)


plt.title('Tumor Response to Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.legend()
plt.grid()

# Save the Figure
plt.savefig('Images/Tumor Response to Treatment.png', dpi=200)
```

**Only "Capomulin" made a positive effect**

## Metastatic Response to Treatment

```python
# Select columns of interest 
metastatic_response = merged_df[['Drug', 'Timepoint', 'Metastatic Sites']]

#Store the Mean Met. Site Data Grouped by Drug and Timepoint 
metastatic_response = pd.DataFrame(metastatic_response.groupby(by=['Drug', 'Timepoint'])['Metastatic Sites'].mean())
metastatic_response.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
metastatic_response['Metastatic Sites SEM'] = metastatic_response['Metastatic Sites'].sem()
metastatic_response.head()
```

```python
# Create pivot two table one with Metastatic Sites and another with Metastatic Sites SEM
metastatic_response_pivot = pd.pivot_table(data=metastatic_response, values='Metastatic Sites', 
                                           columns='Drug', index='Timepoint')
metastatic_response_pivot.head()
```

```python
metastatic_response_sem_pivot = pd.pivot_table(data=metastatic_response, values='Metastatic Sites SEM', index='Timepoint', columns='Drug')
metastatic_response_sem_pivot.head()
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(x=metastatic_response_pivot.index, yerr=metastatic_response_sem_pivot['Capomulin'], 
             y=metastatic_response_pivot['Capomulin'], 
             color='r', marker='o', markersize=5, linestyle='--', linewidth=0.5)
plt.errorbar(x=metastatic_response_pivot.index, yerr=metastatic_response_sem_pivot['Infubinol'], 
             y=metastatic_response_pivot['Infubinol'], 
             color='b', marker='^', markersize=5, linestyle='--', linewidth=0.5)
plt.errorbar(x=metastatic_response_pivot.index, yerr=metastatic_response_sem_pivot['Ketapril'], 
             y=metastatic_response_pivot['Ketapril'],
             color='g', marker='s', markersize=5, linestyle='--', linewidth=0.5)
plt.errorbar(x=metastatic_response_pivot.index, yerr=metastatic_response_sem_pivot['Placebo'],
            y=metastatic_response_pivot['Placebo'],
            color='black', marker='D', markersize=5, linestyle='--', linewidth=0.5)
plt.grid()
plt.legend()
plt.title('Metastatic Spread During Treatment')
plt.xlabel('Treatment Duration (Days)')
plt.ylabel('Metastatic Sites')

# Save the Figure
plt.savefig('Images/Metastatic Spread During Treatment', dpi=200)
```

## Survival Rates

```python
# Select columns of interest
survival_rate = merged_df[['Timepoint', 'Drug', 'Mouse ID']]

# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
survival_rate = survival_rate.set_index(['Timepoint', 'Drug'])
survival_rate.head()

# Add "Mouse Count" column
survival_rate['Mouse Count'] = survival_rate.groupby(['Timepoint', 'Drug'])['Mouse ID'].count()
survival_rate.head()
```

**As time goes on number of "Mouse Id" decrease**

```python
# Create a pivot table to show a change in mouse count over time
survival_rate = pd.pivot_table(data=survival_rate, values='Mouse Count', index='Timepoint', columns='Drug')
survival_rate
```

```python
# Calcutlate percentage change and plot the data
plt.plot((survival_rate['Capomulin']/25)*100, marker='o',linestyle='--', linewidth=0.5, color='r')
plt.plot((survival_rate['Infubinol']/25)*100, marker='^',linestyle='--', linewidth=0.5, color='b')
plt.plot((survival_rate['Ketapril']/25)*100, marker='s',linestyle='--', linewidth=0.5, color='g')
plt.plot((survival_rate['Placebo']/25)*100, marker='D',linestyle='--', linewidth=0.5, color='black')

plt.grid()
plt.title('Survival During Treatment')
plt.ylabel('Survival Rate (%)')
plt.xlabel('Time (Days)')
plt.legend()

# Save the Figure
plt.savefig('Images/Survival During Treatment', dpi=200)

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
tumor_volume_change = ((tumor_response_pivot.iloc[-1]-tumor_response_pivot.iloc[0])/tumor_response_pivot.iloc[0])*100
tumor_volume_change
```

```python
tumor_volume_change.values
```

```python
# Store all Relevant Percent Changes into a Tuple
tumor_volume_change = tumor_volume_change[['Capomulin','Infubinol','Ketapril','Placebo']]
tumor_volume_change.values

# Orient widths. Add labels, tick marks, etc. 
# Use functions to label the percentages of changes
x_axis = tumor_volume_change.index
tick_location = np.arange(len(x_axis))

plt.bar(x=x_axis, height=tumor_volume_change.values, 
       color = ['red' if tumor_volume_change[i]>0 else 'green' for i in np.arange(len(x_axis))])
plt.xticks(tick_location, labels=x_axis)

plt.axhline()
plt.ylim(min(tumor_volume_change)-5, max(tumor_volume_change)+3)
#plt.tick_params(30)
plt.title("Tumor Change Over 45 Day Treatment")
plt.xlabel("Drug Names")
plt.ylabel("% Tumor Volume Change")
plt.grid()

# Save the Figure
plt.savefig('Images/Tumor Change Over 45 Day Treatment', dpi=200)

# Show the Figure
plt.show()
```
